```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
def train_ann(X, y, hidden_neurons=2, learning_rate=0.1, epochs=10000):
    np.random.seed(42)
    input_neurons = X.shape[1]
    output_neurons = 1
    W1 = np.random.rand(input_neurons, hidden_neurons)
    b1 = np.zeros((1, hidden_neurons))
    W2 = np.random.rand(hidden_neurons, output_neurons)
    b2 = np.zeros((1, output_neurons))

    for epoch in range(epochs):
        hidden_input = np.dot(X, W1) + b1
        hidden_output = sigmoid(hidden_input)
        final_input = np.dot(hidden_output, W2) + b2
        final_output = final_input
        loss = np.mean((y - final_output) ** 2)
        error = final_output - y
        dW2 = np.dot(hidden_output.T, error) / X.shape[0]
        db2 = np.mean(error, axis=0)
        hidden_error = np.dot(error, W2.T) * sigmoid_derivative(hidden_output)
        dW1 = np.dot(X.T, hidden_error) / X.shape[0]
        db1 = np.mean(hidden_error, axis=0)
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        if epoch % 1000 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")

    return W1, b1, W2, b2

def predict(X, W1, b1, W2, b2):
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, W2) + b2
    return final_input
train_X = np.array([[0.1, 0.2], [0.2, 0.3], [0.3, 0.4], [0.6, 0.7], [0.7, 0.8], [0.8, 0.9]])
train_y = np.array([[0.3432], [0.3490], [0.3548], [0.3720], [0.3776], [0.3832]])

W1, b1, W2, b2 = train_ann(train_X, train_y)

test_X = np.array([[0.4, 0.5], [0.5, 0.6]])
test_y = np.array([[0.3606], [0.3663]])

train_predictions = predict(train_X, W1, b1, W2, b2)
test_predictions = predict(test_X, W1, b1, W2, b2)

train_mse = np.mean((train_y - train_predictions) ** 2)
test_mse = np.mean((test_y - test_predictions) ** 2)

print(f"Train MSE: {train_mse}")
print(f"Test MSE: {test_mse}")

x1, x2 = map(float, input("Enter x1 and x2: ").split())
user_input = np.array([[x1, x2]])
user_prediction = predict(user_input, W1, b1, W2, b2)
print(f"Predicted Output: {user_prediction[0,0]}")
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
def linear(x):
    return x
def linear_derivative(x):
    return 1
class NeuralNetwork:
    def __init__(self):
        self.weights1 = np.random.uniform(size=(2, 3))
        self.weights2 = np.random.uniform(size=(3, 1))
        self.bias1 = np.zeros((1, 3))
        self.bias2 = np.zeros((1, 1))
        self.learning_rate = 0.1

    def forward(self, X):
        self.input = X
        self.hidden_input = np.dot(X, self.weights1) + self.bias1
        self.hidden_output = sigmoid(self.hidden_input)
        self.output = np.dot(self.hidden_output, self.weights2) + self.bias2
        self.final_output = linear(self.output)
        return self.final_output

    def backward(self, X, y, output):
        self.output_error = y - output
        self.output_delta = self.output_error * linear_derivative(output)

        self.hidden_error = np.dot(self.output_delta, self.weights2.T)
        self.hidden_delta = self.hidden_error * sigmoid_derivative(self.hidden_output)

        self.weights2 += self.learning_rate * np.dot(self.hidden_output.T, self.output_delta)
        self.bias2 += self.learning_rate * np.sum(self.output_delta, axis=0, keepdims=True)
        self.weights1 += self.learning_rate * np.dot(X.T, self.hidden_delta)
        self.bias1 += self.learning_rate * np.sum(self.hidden_delta, axis=0, keepdims=True)

    def train(self, X, y, epochs=1000):
        for _ in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output)

    def predict(self, X):
        return self.forward(X)

train_data = np.array([
    [0.2, 0.1],
    [0.3, 0.2],
    [0.4, 0.3],
    [0.7, 0.6],
    [0.8, 0.7],
    [0.9, 0.8]
])

train_target = np.array([
    [0.3441],
    [0.3500],
    [0.3558],
    [0.3729],
    [0.3785],
    [0.3841]
])

test_data = np.array([
    [0.5, 0.4],
    [0.6, 0.5]
])

test_target = np.array([
    [0.3615],
    [0.3672]
])

nn = NeuralNetwork()
nn.train(train_data, train_target, epochs=1000)
def calculate_mse(predictions, targets):
    return np.mean((predictions - targets) ** 2)
train_predictions = nn.predict(train_data)
train_mse = calculate_mse(train_predictions, train_target)
print(f"Training MSE: {train_mse:.6f}")
test_predictions = nn.predict(test_data)
test_mse = calculate_mse(test_predictions, test_target)
print(f"Testing MSE: {test_mse:.6f}")
def get_user_prediction(nn):
    try:
        x1 = float(input("Enter x1 value: "))
        x2 = float(input("Enter x2 value: "))
        input_data = np.array([[x1, x2]])
        prediction = nn.predict(input_data)
        print(f"Predicted output: {prediction[0][0]:.4f}")
    except ValueError:
        print("Please enter valid numerical values")

print("\nEnter values for prediction:")
get_user_prediction(nn)

print("\nNetwork Architecture:")
print("Input Layer: 2 neurons (x1, x2)")
print("Hidden Layer: 3 neurons (sigmoid activation)")
print("Output Layer: 1 neuron (linear activation)")
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def linear(x):
    return x

def linear_derivative(x):
    return 1

class NeuralNetwork:
    def __init__(self):
        self.w1 = np.random.randn(2, 3) * 0.1
        self.w2 = np.random.randn(3, 1) * 0.1
        self.b1 = np.zeros((1, 3))
        self.b2 = np.zeros((1, 1))

    def forward(self, X):
        self.z1 = np.dot(X, self.w1) + self.b1
        self.a1 = sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.w2) + self.b2
        self.a2 = linear(self.z2)
        return self.a2

    def backward(self, X, y, output, learning_rate=0.1):
        self.error = y - output
        self.delta2 = self.error * linear_derivative(output)
        self.error_hidden = np.dot(self.delta2, self.w2.T)
        self.delta1 = self.error_hidden * sigmoid_derivative(self.a1)
        self.w2 += learning_rate * np.dot(self.a1.T, self.delta2)
        self.b2 += learning_rate * np.sum(self.delta2, axis=0, keepdims=True)
        self.w1 += learning_rate * np.dot(X.T, self.delta1)
        self.b1 += learning_rate * np.sum(self.delta1, axis=0, keepdims=True)

    def train(self, X, y, epochs=1000):
        for _ in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output)

    def calculate_mse(self, X, y):
        predictions = self.forward(X)
        return np.mean((y - predictions) ** 2)

train_X = np.array([[1, 2],
                    [2, 3],
                    [5, 6],
                    [6, 7],
                    [7, 8],
                    [8, 9]])
train_y = np.array([[0.4140],
                    [0.4611],
                    [0.5501],
                    [0.5656],
                    [0.5765],
                    [0.5840]])

test_X = np.array([[3, 4],
                   [4, 5]])
test_y = np.array([[0.4992],
                   [0.5285]])

nn = NeuralNetwork()
print("Initial weights:")
print("W1:", nn.w1)
print("W2:", nn.w2)

nn.train(train_X, train_y, epochs=1000)

train_mse = nn.calculate_mse(train_X, train_y)
test_mse = nn.calculate_mse(test_X, test_y)

print("\nAfter training:")
print("Training MSE:", train_mse)
print("Testing MSE:", test_mse)
print("Final weights:")
print("W1:", nn.w1)
print("W2:", nn.w2)

def predict_output():
    try:
        x1 = float(input("Enter x1: "))
        x2 = float(input("Enter x2: "))
        input_data = np.array([[x1, x2]])
        prediction = nn.forward(input_data)
        print(f"Predicted output: {prediction[0][0]:.4f}")
    except ValueError:
        print("Please enter valid numerical values")

print("\nTest your own input:")
predict_output()
```
